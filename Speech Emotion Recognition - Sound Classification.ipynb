```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
from IPython.display import Audio
import os

# Helper functions for plotting waveplot and spectrogram
def waveplot(data, sampling_rate, emotion):
    """
    Function to plot waveplot for a given emotion.
    """
    plt.figure(figsize=(10, 4))
    librosa.display.waveshow(data, sr=sampling_rate)
    plt.title(f"Waveplot for {emotion.capitalize()}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.show()

def spectogram(data, sampling_rate, emotion):
    """
    Function to plot spectrogram for a given emotion.
    """
    X = librosa.stft(data)
    X_db = librosa.amplitude_to_db(np.abs(X), ref=np.max)
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(X_db, sr=sampling_rate, x_axis='time', y_axis='hz', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title(f"Spectrogram for {emotion.capitalize()}")
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")
    plt.show()

# List of emotions to process
# emotions = ['fear', 'angry', 'disgust', 'neutral', 'sad', 'ps', 'happy']
emotions = os.listdir(r'C:\Users\dhruv\OneDrive\Desktop\Speech\Emotions')

# Iterate over each emotion and process the corresponding audio
for emotion in emotions:
    try:
        print(f"Processing emotion: {emotion}")
        
        # Get the path of the first audio sample for the given emotion
        # path = np.array(df['speech'][df['label'] == emotion])[0]
        path = os.path.join(r'C:\Users\dhruv\OneDrive\Desktop\Speech\Emotions', emotion) 
        for file in os.listdir(path):
            print(file)
               
        
        # Load the audio file
            data, sampling_rate = librosa.load(os.path.join(path, file))
            
            # Print the sampling details
            print(f"Loaded audio file for emotion '{emotion}':")
            print(f"Data: {data[:10]}... (truncated for display)")
            print(f"Sampling rate: {sampling_rate} Hz")
            
            # Plot waveplot
            print(f"Constructing waveplot for label '{emotion.capitalize()}'")
            waveplot(data, sampling_rate, emotion)
            
            # Plot spectrogram
            print(f"Constructing spectrogram for label '{emotion.capitalize()}'")
            spectogram(data, sampling_rate, emotion)
            
            # Play the audio
            display(Audio(path))
            break
            
    except IndexError:
        print(f"No audio file found for emotion '{emotion}'. Skipping.")
    except Exception as e:
        print(f"An error occurred while processing emotion '{emotion}': {e}")
```

## Feature Extraction

```python
# Function to extract MFCC features from an audio file
def extract_mfcc(filename):
    """
    Extracts Mel-Frequency Cepstral Coefficients (MFCC) features from an audio file.
    Args:
        filename (str): Path to the audio file.
    Returns:
        np.array: Mean MFCC features with 40 coefficients.
    """
    try:
        # Load the audio file with librosa (offset and duration to standardize input length)
        y, sr = librosa.load(filename, offset=0.5, duration=3)
        
        # Extract MFCC features with 40 coefficients
        mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sampling_rate, n_mfcc=40).T, axis=0)
        return mfcc
    except FileNotFoundError:
        print(f"Error: File not found for {filename}")
        return np.zeros(40)  # Return zeros for missing files
    except Exception as e:
        print(f"An error occurred while extracting MFCC from {filename}: {e}")
        return np.zeros(40)  # Return zeros for errors

# Example usage assuming 'df' contains a column 'speech' with file paths
try:
    # Extract MFCC features for the first audio file in the dataset (for testing)
    sample_mfcc = extract_mfcc(df['speech'][0])
    print("Sample MFCC extracted:", sample_mfcc)
    
    # Apply the extract_mfcc function to all audio files in the dataset
    print("Extracting MFCC features for all audio files...")
    X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
    
    # Display extracted MFCCs
    print("MFCC features extracted for all files:")
    print(X_mfcc.head())
    
    # Convert the list of MFCC features into a NumPy array
    X = [x for x in X_mfcc]
    X = np.array(X)
    print("Shape of feature array:", X.shape)
    
    # Expand dimensions for compatibility with deep learning models
    X = np.expand_dims(X, -1)  # Adds a channel dimension
    print("Shape after expanding dimensions:", X.shape)
except NameError:
    print("Error: DataFrame 'df' is not defined. Please load the dataset.")
```

```python
from sklearn.preprocessing import OneHotEncoder

# Ensure the DataFrame exists and contains the 'label' column
try:
    if 'label' not in df.columns:
        raise ValueError("The 'label' column is missing in the DataFrame.")
    
    # Initialize the OneHotEncoder
    enc = OneHotEncoder(sparse_output=False)  # Set sparse_output=False to directly get an array
    
    # Perform one-hot encoding on the 'label' column
    y = enc.fit_transform(df[['label']])  # [[label]] ensures it is treated as a 2D array
    
    # Check the shape of the resulting array
    print("Shape of one-hot encoded array:", y.shape)
    
    # Optionally, display the mapping of labels to one-hot encoded vectors
    print("Categories for OneHotEncoder:", enc.categories_)

except NameError:
    print("Error: DataFrame 'df' is not defined. Please load the dataset.")
except ValueError as e:
    print(f"Error: {e}")
```

## Create the LSTM Model

```python
#LSTM- Long Short-Term Memory
# Importing necessary modules from Keras

from keras.models import Sequential  # Used to create a linear stack of layers for the neural network
from keras.layers import Dense, LSTM, Dropout  # Layers used for building the model

# Define the Sequential model
model = Sequential([ #Sequential data is data that is arranged in a sequence, where the elements are dependent on each other. Examples of sequential data include text, audio, video, and time series data.
    
    # LSTM layer:
    # - 256 units: Number of LSTM units (hidden states).
    # - return_sequences=False: Since we are not stacking multiple LSTM layers, we only need the final output.
    # - input_shape=(40,1): Specifies input dimensions, where 40 is the MFCC feature length, and 1 is the number of channels.
    LSTM(256, return_sequences=False, input_shape=(40, 1)),
    
    # Dropout layer:
    # - 0.5: Drops 50% of the units randomly during training to prevent overfitting.
    # Dropout(0.5),
    
    # Dense layer:
    # - 128 units: Number of neurons in this fully connected layer.
    # - activation='relu': Rectified Linear Unit (ReLU) activation function, commonly used for hidden layers.
    Dense(128, activation='relu'),
    
    # Dropout layer:
    # - 0.5: Drops 50% of the units randomly to further reduce overfitting.
    Dropout(0.5),
    
    # Dense layer:
    # - 64 units: Reduces the number of neurons to focus on higher-level abstractions.
    # - activation='relu': ReLU activation again for better gradient flow.
    Dense(64, activation='relu'),
    
    # Dropout layer:
    # - 0.5: Drops 50% of the units again.
    # Dropout(0.5),
    
    # Output layer:
    # - 7 units: Number of output neurons, corresponding to 7 emotion categories.
    # - activation='softmax': Softmax activation for multi-class classification, outputs probabilities for each class.
    Dense(7, activation='softmax')
])

# Compile the model:
# - loss='categorical_crossentropy': Loss function for multi-class classification.
# - optimizer='adam': Adaptive Moment Estimation optimizer, efficient for deep learning models.
# - metrics=['accuracy']: Tracks accuracy as a metric during training and evaluation.
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display the model's architecture and parameters
model.summary()  # Prints a summary of the model, showing layer types, output shapes, and parameter counts
```

```python
# Train the model
history = model.fit(X, y, validation_split=0.1, epochs=50, batch_size=32)
```

## Plot the results

```python
# Assuming the model has been trained and the training history is stored in the 'history' object

# Define the range of epochs for plotting
epochs = list(range(50))  # A list of epoch numbers from 0 to 49

# Extract accuracy values from the training history
acc = history.history['accuracy']  # Training accuracy over epochs
val_acc = history.history['val_accuracy']  # Validation accuracy over epochs

# Plot training and validation accuracy
plt.plot(epochs, acc, label='train accuracy')  # Line plot for training accuracy
plt.plot(epochs, val_acc, label='val accuracy')  # Line plot for validation accuracy
plt.xlabel('epochs')  # Label for x-axis
plt.ylabel('accuracy')  # Label for y-axis
plt.legend()  # Display legend to distinguish lines
plt.title('Training and Validation Accuracy')  # Title of the plot
plt.show()  # Display the plot

# Extract loss values from the training history
loss = history.history['loss']  # Training loss over epochs
val_loss = history.history['val_loss']  # Validation loss over epochs

# Plot training and validation loss
plt.plot(epochs, loss, label='train loss')  # Line plot for training loss
plt.plot(epochs, val_loss, label='val loss')  # Line plot for validation loss
plt.xlabel('epochs')  # Label for x-axis
plt.ylabel('loss')  # Label for y-axis
plt.legend()  # Display legend to distinguish lines
plt.title('Training and Validation Loss')  # Title of the plot
plt.show()  # Display the plot
```
